```python
import numpy as np
target1 = np.array(target, dtype=np.int64)
```

```python
X["target"] = target1
X.head()
```

```python
X.info()
```

```python
X_corr = X.corr()
X_corr.info()
```

```python
import matplotlib
import matplotlib.image as img
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
%matplotlib inline
```

```python
plt.figure(figsize = (15,10))

sns.set(font_scale=1.4)

corr_matrix = X_corr

corr_matrix = np.round(corr_matrix, 2)
corr_matrix[np.abs(corr_matrix) <= 0.5] = 0

sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm')

plt.title('Correlation matrix')
plt.show()
```

```python
high_corr = np.round(X_corr, 2)
high_corr=high_corr.loc[(np.abs(X_corr['target']) > 0.5)&(X_corr.columns != 'target'), X_corr.index != 'target'].index
high_corr
```

```python
X.drop('target', axis=1, inplace=True)
X.head()
```

```python
for name in high_corr:
    X[f'{name}_2']=X[name]**2
X.head()        
```
